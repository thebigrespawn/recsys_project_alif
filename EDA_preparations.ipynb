```python
#upload sheets with businessess and reviews
file_path_business = './data/yelp_dataset/yelp_academic_dataset_business.json'

business_df = pd.read_json(file_path_business, lines=True)

business_df = business_df[business_df['categories'].str.contains("Restaurants", na=False)]

business_ids_set = set(business_df['business_id'])

file_path_review = './data/yelp_dataset/yelp_academic_dataset_review.json'

filtered_reviews = []
#chunk loading needed as json is too heavy
chunk_size = 100000

for chunk in pd.read_json(file_path_review, lines=True, chunksize=chunk_size)
    filtered_chunk = chunk[chunk['business_id'].isin(business_ids_set)]
    filtered_reviews.append(filtered_chunk)

review_df = pd.concat(filtered_reviews, ignore_index=True)
```

```python
#import all the places they visited
df = pd.merge(sample_reviews[['business_id', 'date']], business_df, how='inner')[['latitude', 'longitude', 'date']]
df
```

```python
#try to cluster the places and find the spot where user likes to be the most

import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from geopy.distance import geodesic
from geopy.point import Point
import folium

### step 1: cluster cafes using DBSCAN
# convert latitude and longitude to radians for haversine metric to calculate real distances from lat. and long.
coords = np.radians(df[['latitude', 'longitude']].values)

# DBSCAN with eps=5km converted to radians (eps is ~5/6371 for Earth's radius in km)
# I have supposed that preferable area may be around 5km in radius considering the specifics of US where everone has cars
earth_radius_km = 6371  # Earth's radius in kilometers
eps_rad = 5 / earth_radius_km

# apply DBSCAN clustering
db = DBSCAN(eps=eps_rad, min_samples=3, metric='haversine').fit(coords)
df['cluster'] = db.labels_

### step 2: find the most dense cluster
# identify the largest cluster (excluding noise points: cluster = -1)
most_dense_cluster = df['cluster'].value_counts().idxmax()
cluster_points = df[df['cluster'] == most_dense_cluster]

# calculate the geographic center of the most dense cluster
cluster_center = cluster_points[['latitude', 'longitude']].mean()

### step 3: visualize the clusters and the center
# create a map centered at the cluster center
m = folium.Map(location=[cluster_center['latitude'], cluster_center['longitude']], zoom_start=10)


for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=4,
        color='blue' if row['cluster'] == most_dense_cluster else 'gray',
        fill=True,
        fill_opacity=0.7
    ).add_to(m)


folium.Marker(
    location=[cluster_center['latitude'], cluster_center['longitude']],
    popup="Center of Most Dense Cluster",
    icon=folium.Icon(color='red', icon="info-sign")
).add_to(m)


# display results
print(f"Center of Most Dense Cluster: {cluster_center}")
```

```python
# preprocessing attributes and categories
import re
import ast
import json
import pandas as pd

def try_parse_dict(value):
    if isinstance(value, dict):
        return value
    if value is None or (isinstance(value, float) and pd.isna(value)) or str(value).strip() == '':
        return {}
    #unify format as lowercase
    s = str(value).strip()
    s = re.sub(r"\btrue\b", "True", s, flags=re.IGNORECASE)
    s = re.sub(r"\bfalse\b", "False", s, flags=re.IGNORECASE)
    s = re.sub(r"\bnone\b", "None", s, flags=re.IGNORECASE)
    try:
        parsed = ast.literal_eval(s)
        return parsed if isinstance(parsed, dict) else {}
    except:
        json_str = s.replace("'", '"')
        try:
            parsed_json = json.loads(json_str)
            return parsed_json if isinstance(parsed_json, dict) else {}
        except:
            return {}
#flatten nested dicts, keep hierarchies
def flatten_nested_dict(d):
    flattened = {}
    for key, value in d.items():
        if isinstance(value, str) and value.strip().startswith("{") and value.strip().endswith("}"):
            nested = try_parse_dict(value)
            for nk, nv in nested.items():
                flattened[f"{key}_{nk}"] = nv
        else:
            flattened[key] = value
    return flattened

#leave attribute only if it is true
def clean_attributes(attributes):
    cleaned = {}
    for k, v in attributes.items():
        if v is None or v is False:
            continue
        if isinstance(v, str) and v.strip().lower() in ['false', 'none', '']:
            continue
        cleaned[k] = v
    return cleaned

def parse_and_flatten_attributes(attr_value):
    parsed = try_parse_dict(attr_value)
    flattened = flatten_nested_dict(parsed)
    cleaned = clean_attributes(flattened)
    return cleaned

def transform_key_value(k, v):
    k = k.lower()
    v_str = str(v).lower().strip()

    # Remove 'u'' prefix if present
    if v_str.startswith("u'"):
        v_str = v_str[2:].strip()

    # skip parking handling here (we do it in combine_features)
    if k.startswith("businessparking_"):
        return ""

    # handle alcohol special case
    if k == 'alcohol':
        if v_str == "'full_bar'":
            # If full_bar is present, also mention beer_and_wine
            return 'alcohol_beer_and_wine alcohol_full_bar'
    
    # if value is none or no, skip
    if v_str in ("none", 'no'):
        return ''
    # if value is true or yes, just return the key
    if v_str in ('true', 'yes'):
        return k
    # otherwise return key_value where it is not binary
    return f"{k}_{v_str}".replace(' ', '')

def combine_features(row):
    # process categories: split by comma, strip, remove spaces
    categories_str = row['categories'].lower() if pd.notna(row['categories']) else ''
    categories_list = [cat.strip() for cat in categories_str.split(',')]
    processed_cats = [cat.replace(' ', '') for cat in categories_list]
    categories_str = ' '.join(processed_cats)

    attrs = row['flattened_attributes']

    # determine parking presence, concat all parking options
    parking_attrs = {k.lower(): str(v).lower().strip()
                     for k, v in attrs.items()
                     if k.lower().startswith("businessparking_")}

    true_count = sum(1 for val in parking_attrs.values() if val == "true")
    validated_true = (parking_attrs.get('businessparking_validated') == 'true')

    # parking present logic
    # if no parking attributes are True -> no parking
    # if only validated is True (and nothing else) -> no parking
    # otherwise parking is present
    if true_count == 0:
        parking_present = False
    elif true_count == 1 and validated_true:
        parking_present = False
    else:
        parking_present = (true_count > 0)

    transformed_attrs = []
    for k, v in attrs.items():
        token = transform_key_value(k, v)
        if token:
            transformed_attrs.append(token)

    # if parking is present, add 'parking' once to keep it clean
    if parking_present:
        transformed_attrs.append("parking")

    attributes_str = ' '.join(transformed_attrs)
    
    combined = f"{categories_str} {attributes_str}".strip()
    
    # final cleanup
    combined = re.sub(r'[^\w\s]', '', combined)    # Remove punctuation
    combined = re.sub(r'_+', '_', combined)         # Collapse multiple underscores
    combined = re.sub(r'\s+', ' ', combined)        # Collapse multiple spaces
    combined = combined.strip()

    return combined

#run example
business_df['flattened_attributes'] = business_df['attributes'].apply(parse_and_flatten_attributes)
business_df['combined_features'] = business_df.apply(combine_features, axis=1)
print("Sample combined_features:\n", business_df['combined_features'].iloc[9])
```

```python
import pandas as pd
from collections import Counter

# Example:
# reviews_df columns: user_id, business_id, date (plus possibly others)
# business_df columns: business_id, combined_features

# Merge reviews with business info
reviews_with_business = pd.merge(review_df, business_df, on='business_id', how='left')

# Sort by user and date so we can check continuity of preferences
reviews_with_business = reviews_with_business.sort_values(by=['user_id', 'date'])

# Tokenize combined_features into lists of tokens
reviews_with_business['tokens'] = reviews_with_business['combined_features'].apply(
    lambda x: x.split() if isinstance(x, str) else []
)

def analyze_user_preferences(user_id, df):
    """
    For a given user:
    1. Calculate the frequency of each token across all visited businesses.
    2. Calculate how 'continuous' the user's preference is for those tokens by checking consecutive visits.
    """
    user_data = df[df['user_id'] == user_id].copy()
    if user_data.empty:
        return [], []

    user_data = user_data.sort_values('date')
    visits_tokens = user_data['tokens'].tolist()

    # Count overall frequency of tokens
    all_tokens = [token for visit in visits_tokens for token in visit]
    freq = Counter(all_tokens)

    # Calculate continuity: check pairs of consecutive visits and count how many tokens appear in both
    continuity_count = Counter()
    for i in range(len(visits_tokens)-1):
        current_visit = set(visits_tokens[i])
        next_visit = set(visits_tokens[i+1])
        common = current_visit.intersection(next_visit)
        for token in common:
            continuity_count[token] += 1
    
    max_pairs = len(visits_tokens) - 1
    continuity_score = {token: continuity_count[token] / max_pairs if max_pairs > 0 else 0
                        for token in freq}

    # Top frequent tokens
    top_frequent = freq.most_common(100)

    # Top continuous tokens
    top_continuous = sorted(continuity_score.items(), key=lambda x: x[1], reverse=True)[:100]

    return top_frequent, top_continuous

# Example usage
some_user_id = '6I1jouo2T1R-X2ykY2T65w'
top_frequent, top_continuous = analyze_user_preferences(some_user_id, reviews_with_business)

print(f"User {some_user_id} Top Frequent Tokens:")
for token, count in top_frequent:
    print(f"{token}: {count}")

print(f"\nUser {some_user_id} Top Continuous Tokens:")
for token, score in top_continuous:
    print(f"{token}: continuity_score={score:.2f}")
```

```python
import json
from collections import Counter

def count_business_attributes(file_path):
    """
    Count the number of attributes for businesses in the Yelp dataset.

    :param file_path: Path to the Yelp business dataset JSON file
    :return: Dictionary with counts of attributes across all businesses
    """
    attribute_counter = Counter()

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                business = json.loads(line)
                if 'attributes' in business and isinstance(business['attributes'], dict):
                    attribute_counter.update(business['attributes'].keys())
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return {}
    except json.JSONDecodeError:
        print(f"Error decoding JSON in file: {file_path}")
        return {}

    return dict(attribute_counter)

# Example usage
file_path = "./data/yelp_dataset/yelp_academic_dataset_business.json"  # Replace with your file path
attribute_counts = count_business_attributes(file_path)

# Display attribute counts
for attribute, count in sorted(attribute_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{attribute}: {count}")
```
